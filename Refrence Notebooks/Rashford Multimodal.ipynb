```python
import re

def clean_transcript(text):
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    text = re.sub(r'\bi\b', 'I', text)  # Capitalize standalone 'i'
    text = text.strip()
    
    # Split sentences after trigger words without lookbehind
    split_pattern = r'\b(?:yeah|go|get|year|won|want|them)\b\s+'
    sentences = re.split(split_pattern, text)
    sentences = [s.strip().capitalize() + '.' for s in sentences if len(s.strip()) > 0]
    return "\n".join(sentences)

# Example usage:
print("🧼 Cleaned Transcript:\n")
print(clean_transcript(transcript))
```

```python
import re
import nltk
from fuzzywuzzy import fuzz
nltk.download('punkt')

def normalize(text):
    return re.sub(r"[^\w\s]", "", text.lower())

def align_sentences_fuzzy(cleaned_transcript, word_results, threshold=85):
    sentences = nltk.sent_tokenize(cleaned_transcript)
    vosk_words = [normalize(w['word']) for w in word_results]
    timestamps = word_results

    aligned = []
    index = 0

    for sentence in sentences:
        norm_sentence = normalize(sentence)
        sentence_words = norm_sentence.split()
        best_score = 0
        best_i = None

        for i in range(index, len(vosk_words) - len(sentence_words) + 1):
            window = vosk_words[i:i + len(sentence_words)]
            window_str = " ".join(window)
            score = fuzz.partial_ratio(" ".join(sentence_words), window_str)
            if score > best_score and score >= threshold:
                best_score = score
                best_i = i

        if best_i is not None:
            start_time = timestamps[best_i]['start']
            end_time = timestamps[best_i + len(sentence_words) - 1]['end']
            index = best_i + len(sentence_words)
        else:
            start_time = None
            end_time = None

        aligned.append({
            'sentence': sentence,
            'start_time': start_time,
            'end_time': end_time,
            'match_score': best_score
        })

    return aligned
```

```python
import wave
import json
import re
import nltk
from vosk import Model, KaldiRecognizer
from difflib import SequenceMatcher
import pandas as pd
import os

nltk.download('punkt')

# --- Load Vosk model and audio file ---
model_path = "C:\\Users\\91722\\Downloads\\vosk-model-small-en-us-0.15\\vosk-model-small-en-us-0.15"
audio_path = "C:\\Users\\91722\\Desktop\\SPJIMR Mumbai\\My Courses\\Text_Image_Video Analytics using AI\\PPTs\\Session 12\\rashford_converted_for_vosk.wav"


model = Model(model_path)
wf = wave.open(audio_path, "rb")

if wf.getnchannels() != 1 or wf.getsampwidth() != 2 or wf.getframerate() != 16000:
    raise ValueError("Audio must be WAV format mono PCM with 16kHz")

# --- Recognize words with timestamps ---
rec = KaldiRecognizer(model, wf.getframerate())
rec.SetWords(True)
word_results = []

while True:
    data = wf.readframes(4000)
    if len(data) == 0:
        break
    if rec.AcceptWaveform(data):
        part_result = json.loads(rec.Result())
        if "result" in part_result:
            word_results.extend(part_result["result"])

final_result = json.loads(rec.FinalResult())
if "result" in final_result:
    word_results.extend(final_result["result"])

# --- Clean and segment transcript into sentences ---
def clean_and_segment_transcript(text):
    text = re.sub(r'\s+', ' ', text.strip())
    sentences = nltk.sent_tokenize(text)
    return sentences

cleaned_transcript = clean_transcript (transcript)
sentences = clean_and_segment_transcript(cleaned_transcript)

# --- Align sentences with timestamps using fuzzy match ---
aligned = []
words = [w['word'].lower() for w in word_results]

for sentence in sentences:
    sentence_words = sentence.lower().split()
    max_ratio = 0
    best_start, best_end = None, None

    for i in range(len(words) - len(sentence_words) + 1):
        window = words[i:i + len(sentence_words)]
        ratio = SequenceMatcher(None, sentence_words, window).ratio()
        if ratio > max_ratio:
            max_ratio = ratio
            best_start = i
            best_end = i + len(sentence_words)

    if best_start is not None and best_end <= len(word_results):
        start_time = word_results[best_start]["start"]
        end_time = word_results[best_end - 1]["end"]
    else:
        start_time = end_time = None

    aligned.append((start_time, end_time, max_ratio, sentence))

df = pd.DataFrame(aligned, columns=["start", "end", "score", "sentence"])
print(df.head())
```

```python
import cv2

video_path = "C:\\Users\\91722\\Desktop\\SPJIMR Mumbai\\My Courses\\Text_Image_Video Analytics using AI\\PPTs\\Session 12\\rashford_video.mp4"
cap = cv2.VideoCapture(video_path)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
frame_count = 0
face_times = []

fps = cap.get(cv2.CAP_PROP_FPS)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    time_in_sec = frame_count / fps

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    if len(faces) > 0:
        face_times.append(time_in_sec)

cap.release()

print(f"🙎 Face appeared in {len(face_times)} frames")
print(f"🕒 Example timestamps: {face_times[:5]}")
```

```python
import pandas as pd
import difflib

# Assuming 'cleaned_transcript' is your final transcript string
# And 'results' is your list of word-level Vosk results
# Each result: {'word': str, 'start': float, 'end': float, 'conf': float}

# Split cleaned transcript into words and group into "pseudo-sentences" (e.g., every ~20 words)
words = cleaned_transcript.strip().split()
group_size = 20
sentences = [" ".join(words[i:i+group_size]) for i in range(0, len(words), group_size)]

# Create a dataframe from Vosk results
df_words = pd.DataFrame(word_results)

aligned_sentences = []

for sent in sentences:
    sent_words = sent.lower().split()
    candidate_idxs = df_words['word'].str.lower().tolist()

    # Match first and last word using closest match
    try:
        start_word = difflib.get_close_matches(sent_words[0], candidate_idxs, n=1)[0]
        end_word = difflib.get_close_matches(sent_words[-1], candidate_idxs[::-1], n=1)[0]

        start_idx = candidate_idxs.index(start_word)
        end_idx = len(candidate_idxs) - 1 - candidate_idxs[::-1].index(end_word)

        start_time = df_words.iloc[start_idx]["start"]
        end_time = df_words.iloc[end_idx]["end"]

        aligned_sentences.append({
            "start": float(start_time),
            "end": float(end_time),
            "sentence": sent
        })

    except Exception as e:
        # Couldn't align this sentence
        aligned_sentences.append({
            "start": None,
            "end": None,
            "sentence": sent
        })
```

```python
# Let's align face presence with spoken sentences
multi_modal_events = []

for segment in aligned_sentences:
    for ts in face_times:
        if segment["start"] <= ts <= segment["end"]:
            multi_modal_events.append({
                "start": segment["start"],
                "end": segment["end"],
                "sentence": segment["sentence"],
                "face_time": ts
            })
            break  # Avoid duplicating for multiple face timestamps in same segment

# Display results
for event in multi_modal_events:
    print(f"[{event['start']} - {event['end']}] 🧑‍🗣️ Face Visible → {event['sentence']}")
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# 📍 Audio file path
audio_path = "rashford_converted_for_vosk.wav"

# ✅ Load audio
y, sr = librosa.load(audio_path, sr=16000)

# 🎚️ Calculate short-time RMS energy
frame_length = 2048
hop_length = 512
rms = librosa.feature.rms(y=y, frame_length=frame_length, hop_length=hop_length)[0]
times = librosa.frames_to_time(range(len(rms)), sr=sr, hop_length=hop_length)

# 🧑‍🦱 Face timestamps (in seconds)

# Convert face_times to set for fast lookup
face_set = set(round(ts, 2) for ts in face_times)

# 📊 Plotting
plt.figure(figsize=(16, 6))

# 🌈 Color voice segments based on face visibility
for i, (t, energy) in enumerate(zip(times, rms)):
    t_rounded = round(t, 2)
    color = 'green' if t_rounded in face_set else 'skyblue'
    plt.plot(t, energy, 'o', color=color, markersize=4)

plt.xlabel("Time (seconds)")
plt.ylabel("Voice Energy (RMS)")
plt.title("🎙️ Voice Modulation vs 👤 Face Presence")
plt.grid(True)

from matplotlib.patches import Patch

# Create a list of legend handles
legend_elements = [
    Patch(color='green', label='Face Visible'),
    Patch(color='skyblue', label='Face Not Visible')
]

plt.legend(handles=legend_elements)
```
